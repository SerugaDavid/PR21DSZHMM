# Razvoj in vpliv različnih faktorjev na epidemijo Korona-virusa v Sloveniji.

## Uvod

V tej nalogi smo se kot skupina odločili, da bomo raziskovali področje Korona-virusa v Sloveniji. Postavili smo si neka glavna vprašanja,na katera smo s časom poskušali najti odgovore.
V sklopu naloge smo malo pregledali regije, starostne skupine, cepljenost in odvisnot določenih dogodkov in socialnega življenja na vpliv Korona-virusa.

## Glavna vprašanja

### 1. Kakšne so razlike med regijami in starostnimi skupinami?

Najprej nas je zanimalo, če se število okužb razlikuje glede na regijo.

V ta namen smo pripravili nekaj vizualizacij:


<img src="img/hist_regije_cepljenosti.png">

Histogram predstavlja delež precepljenosti po regijah.

Iz zgornjega histograma lahko razberemo, da se v izbranih različnih regijah precepljost praktično ne razlikuje.

Nato smo po regijah primerjali tudi število okužb. Zanimalo nas je, če so nekatere regije bile bolj izpostavljene Covid okužbam, kot druge.

<img src="img/aktivne_okuzbe_po_regijah_dele%C5%BE.png">

Zgornji graf nam predstavlja delež okuženih ljudi s Covidom v posameznih regijah po datumih. V najviški točki vidimo, da je Podravska dosegla okoli 10% okuženosti populacije. Obalno-krašla je v primerjavi z ostalimi regijami imela konstantno nižjo okuženost populacije.

Seveda pa, so razlike med regijami minimalne in zanemarljive.
Iz tega sklepamo, da se je Covid po Sloveniji širil enakomerno in neodvisno od regije.



#### Kaj pa po starosti?



Podobno kot pri regijah nas je zanimalo ali starostna skupina vpliva na verjetnost, da človek zboli s Covid boleznijo. Zato smo podatke o okužbah razdelili v 5 starostnih skupin. Pri tem pa smo tudi upoštevali število ljudi v posamezni starostni skupini. 

<img src="img/aktivni_po_starosti.png">

Na grafu je razvidno, da je bil največji delež aktivnih okužencev zaznan v najstarejši starostni skupini. To je zelo smiselno, saj so tudi najbolj rizična skupina. Zanimivo pa je, da so kasneje v pandemiji starejši ljudje imeli dokaj nizek procent okuženosti. V jeseni 2021 lahko na grafu vidimo sunek v okužbah pri mlajših generacijah.

Ena od možnih razlag je, da se je previdnost starejših povečala in mlajših zmanjšala.

### 2. Kakšne so razlike med cepljenimi in ne cepljenimi?

Zanimala nas je vplivnost cepiv na število okužb.

Najprej smo ceplejenja z enim odmirkom primerjali po starostnih skupinah. V ta namen smo pripravili graf, kjer smo za vsako starostno skupino primerjali delež precepljenosti posamezne starostne skupine. 

<img src="img/cepljeni%20po%20starosti%20relativno.png">

Na grafu je lepo razviden načrt cepljenja: Najprej pridejo na vrsto starejši, nato pa mlajši. Na začetku leta 2021 se vidi skoraj navpičen vzpon cepljenosti starejših starostnih skupin. Vidimo tudi, da so se mlajše starostne skupine šle cepit kasneje in so tudi dosegle manjši delež precepljenosti. 

<img src="img/hist_po_starosti.png">

Na histogramu lahko vidimo kolikšen delež določene starostne skupine je cepljen vsaj enkrat. 
Lepo vidimo, da mlajša, kot je starostna skupina, manjša je precepljenost te skupine. To je smiselno, saj so starejši bolj nagnjeni k bolezni.


Da pa lahko pogledamo ali je cepljenje pripomoglo k zmanjšanju okužb, moramo primerjati podatke.




<img src="img/relativno_vpliv_cepljenost_na_%C5%A1tevilo_aktivnih_oku%C5%BEencev.png">


Na tem grafu vidimo, da se število aktivnih okužencev (modra krivulja) zmanjša po prvem valu množičnega cepljenja s prvim in drugim odmirkom. Tik pred začetkom distribucije tretjega odmirka je število okužb vendarle naraščalo. Stanje se je takoj po tem sicer izboljšalo, a ne za dolgo, saj vidimo v januarju 2022 največjo rast števila okužencev. Sklepamo, da ta špica ni povezana s cepivi, ampak da so na to vplivale drugi faktorji (npr. novoletni prazniki)

### 3. Kakšen je vpliv socialnega življenja ter vladnih ukrepov?
 
 Do te točke smo že odkrili nekaj dejavnikov, ki so potencialno vplivali na Covid stanje. Vendarle ostajajo določene skrivnosti. Naša zadnja raziskava bo poskušala ugotoviti kako je na število okužb vplivalo socialno življenje. Spremljali bomo tudi vladne ukrepe, saj so ti imeli velik vpliv na življenje v času epidemije.


<img src="img/delez%20pozit%20testov%20s%20faznimi%20prelomnicami.png">


Na grafu zgoraj vidimo glavne prelomnice v ukrepanju vlade. Zelene črte predstavljajo sproščanje ukrepov, rdeče pa zaostrovanje ukrepov. Sklepamo, da vladni ukrepi niso bistveno spreminjali število obolelih, ampak so se bolj prilagajali stanju v Sloveniji ter milili hujše posledice. Menimo, da so za naraščanje in padanje aktivnih Covid okužencev krivi drugi družbeni dejavniki.

Če pogledamo večje praznike in določene prelomnice, ki jih imamo v Sloveniji (novo leto, velika noč, praznik dela, binkošti, začetek šole, začetek študijskega leta, dan spomina na mrtve in božič), lahko vidimo določene korelacije. V začetku leta 2020 imamo še premalo okuženih, da bi lahko videli, kako prazniki vplivajo na našo epidemijo. Situacija se začne zapletati okoli začetka šole oz. začetka novega študijskega leta. Številke pa prav tako narastejo v istem obdobju leta 2021. In tako kot v letu 2020, tudi 2021 se graf zravna v času dneva spomina na mrtve. Seveda, kot pričakovano se primeri dvignejo za ostale velike praznike, ki so velika noč, božič in novo leto. Ostali prazniki kot zgleda nimajo nekega vpliva na epidemijo.

<img src="img/st_aktivnih_in_prazniki.png">

No sedaj, pa lahko pogledamo še smrtnost. Za to imamo še podatke iz prejšnjih let, tako da lahko podatke iz epidemije preverjamo z normalno situacijo.

<img src="img/smrti_letni_casi.png">

Kot vidimo imamo zanimivo vsako leto porast smrti v zimskem času, ki pa se kasneje umiri s poletjem. Na grafu rdeča črtkana navpična črta predstavlja prvo potrjeno okužbo s Korona-virusom. In od takrat naprej vidimo, da je graf za pomlad in poletje precej podoben tem v prejšnjih letih, medtem ko številke postanejo dosti večje jeseni in pozimi. Vrh porasti smrti je za 75% višji kot sezono prej (okoli novega leta 2020). Prav tako lahko vidimo porast eno sezono kasneje (okoli novega leta 2022).

Opazi se tudi, da sta se med epidemijo špici porasta smrti zamaknili v levo. Do sedaj smo vedno imeli vrh po novem letu, zdaj pa ga imamo pred novim letom. Lahko tudi vidimo, da v drugem valu smrti med epidemijo je ta manjši od prvega, čeprav imamo večje število okuženih in prav tako večji delež pozitivnih testov. To nam na nek način govori, da se je zmanjšala moč Korona-virusa. Pa naj bo to zaradi cepiv ali šibkejših različic.
